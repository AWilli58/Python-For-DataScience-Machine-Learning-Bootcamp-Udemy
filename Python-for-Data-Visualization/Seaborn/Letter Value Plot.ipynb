# Seaborn.lvplot()#

*seaborn.lvplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, width=0.8, dodge=True, k_depth='proportion', linewidth=None, scale='exponential', outlier_prop=None, ax=None, **kwargs)

Letter value (LV) plots are non-parametric estimates of the distribution of a dataset, similar to boxplots. LV plots are also similar to violin plots but without the need to fit a kernel density estimate. Thus, LV plots are fast to generate, directly interpretable in terms of the distribution of data, and easy to understand.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
sns.set(context='notebook',palette='Set1',style='whitegrid')
```

```python
plot=sns.distplot(tips['total_bill'],kde=False,color='blue')
```

Try to correlate the distribution of total Bill with the above distribution and the Letter value plot below. 

```python
plot=sns.lvplot(data=tips)
```

Plotting a  Letter Value plot for Total Bill.

```python
plot=sns.lvplot(y='total_bill',data=tips).set(ylabel='Total Bill')
```

Now, lets try to visualize the distribution on a category variable 'Sex'.

```python
plot =sns.lvplot(x="sex", y="total_bill", data=tips).set(ylabel='Total Bill',xlabel='Sex')
plot = sns.stripplot(x='sex',y='total_bill',data=tips,jitter=True,color='grey')
```

Further drill it down based on the **time ('Lunch','Dinner')** and also reorder the '**Male**' and '**Female**' plots. One can control the position of the legend using **legend(loc='')**.
*loc* can take the values of combination between **'upper/bottom' & 'left/right'**. Override the Palette '**Set1**' that is set initially with '**coolwarm**'.

```python
plot =sns.lvplot(x="sex", y="total_bill", data=tips,hue='time',order=['Female','Male'],palette='coolwarm').legend(loc='upper left')
```

Use factorplot() to combine a lvplot() and a FacetGrid. This allows grouping within additional categorical variables. Using factorplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets.

```python
plot = sns.factorplot(x="sex", y="total_bill", hue="time", col="smoker", data=tips, kind="lv",size=4)
```
